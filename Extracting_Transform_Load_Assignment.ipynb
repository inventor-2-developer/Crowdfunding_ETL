```python
import pandas as pd
import re
import csv
```

```python
dataset=pd.read_excel("contacts.xlsx")
dataset_two=pd.read_excel("crowdfunding.xlsx")
dataset_three=pd.read_csv("category.csv")
dataset_four=pd.read_csv("subcategory.csv")
```

```python
dataset.head()
```

```python
#-------------------------------------------Option 2 regular expressions------------------------------------------------#


contact_df = pd.DataFrame(columns=['contact_id','first_name','last_name','email'])
contact_df['contact_id']=contact_df['contact_id'].astype(int)                                   #converting string to int

for index in range(len(dataset)): 
    step_through_row= dataset.iloc[index] 
    filtered_row=list(filter(re.compile(".*{").match,step_through_row))                         # converting dictionary to list
    if filtered_row!=[]:                                                                        # getting rid of the empty lists
        filtered_row_string=''.join(map(str,filtered_row))
        contact_df.loc[index-3,['contact_id']]=re.findall(r'(\d\d\d\d)', filtered_row_string)  # creating df & offsetting index
        contact_df.loc[index-3,['first_name']]=re.findall(r'([A-Z]\w{0,}\s)', filtered_row_string)  
        contact_df.loc[index-3,['last_name']]=re.findall(r'(\s[A-Z]\w{0,})', filtered_row_string) 
        contact_df.loc[index-3,['email']]=re.findall('\w\S+@\S+.{3,4}\w',filtered_row_string) 
```

```python
merge_df=pd.merge(contact_df, dataset_two, on="contact_id")  # joining the dataframes 
merge_df
```

```python
merge_two_df = pd.DataFrame(columns=['category','sub-category'])

for index in range(len(merge_df['category & sub-category'])): 
     merge_two_df.loc[index,['category']]=(merge_df['category & sub-category'][index])
     merge_two_df.loc[index,['sub-category']]=merge_df['category & sub-category'][index]
#    print(values)
    
    
# for index in range(len(merge_df['category & sub-category'])):
#         print(merge_df['category & sub-category'][index])
```
